```python
import pandas as pd
import ast
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('tmdb_5000_movies.csv')
```

```python
df.head()
```

```python
dfc=pd.read_csv('tmdb_5000_credits.csv')
```

```python
dfc.head()
```

```python
films=df.merge(dfc,on='title')
```

```python
films.drop('id',axis=1,inplace=True)
```

```python
films.head()
```

```python
films.columns
```

```python
#'title', 'overview', 'genres', 'keywords', 'cast', and 'crew'
```

```python
movie=films[["movie_id",'title','overview','genres','keywords','cast','crew']]
```

```python
movie
```

```python
movie.isnull().sum()
```

```python
movie['overview'].isna().sum()
```

```python
movie.dropna(inplace=True)
```

```python
movie.isna().sum()
```

```python
movie.info()
```

```python
movie
```

```python
movie['genres'].iloc[0]
```

```python
def gen(obj):
    return [i['name'] for i in obj]
```

```python
movie['genres']=movie['genres'].apply(ast.literal_eval).apply(gen)
```

```python
movie
```

```python
movie['keywords'].iloc[0]
```

```python
def key(obj):
    return [i['name'] for i in obj]
```

```python
movie['keywords']=movie['keywords'].apply(ast.literal_eval).apply(key)
```

```python
movie
```

```python
movie['cast'].iloc[0]
```

```python
def cast(obj):
    return [i['name'] for i in obj[:3]]
```

```python
movie['cast']=movie['cast'].apply(ast.literal_eval).apply(cast)
```

```python
movie
```

```python
movie['crew'].iloc[1]
```

```python
def dir(obj):
    return [i['name'] for i in obj if i.get('job') == 'Director']
```

```python
movie['crew']=movie['crew'].apply(ast.literal_eval).apply(dir)
```

```python
movie
```

```python
movie['overview'][0]
```

```python
movie['overview']=movie['overview'].apply(lambda x:x.split()) ## converting into list to concatenate with other lists
```

```python
movie
```

```python
movie['genres']=movie['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movie['keywords']=movie['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
movie['cast']=movie['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movie['crew']=movie['crew'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movie
```

```python
movie['tags'] =  movie['overview'] + movie['genres'] + movie['keywords'] + movie['cast'] + movie['crew']
```

```python
movie
```

```python
film = movie[['movie_id',"title",'tags']]
```

```python
film
```

```python
film['tags']=film['tags'].apply(lambda x:" ".join(x))
```

```python
film
```

```python
film['tags']=film['tags'].apply(lambda x:x.lower())
```

```python
film['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(film['tags']).toarray()
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
similarity
```

```python
film[film['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(cinema):
    index = film[film['title'] == cinema].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(film.iloc[i[0]].title)
```

```python
recommend('The Dark Knight Rises')
```

```python
import pickle
pickle.dump(film,open('film.pkl','wb'))
pickle.dump(similarity,open('similarity.pkl','wb'))
```
